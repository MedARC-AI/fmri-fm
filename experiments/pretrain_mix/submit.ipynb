```python
from lightning_sdk import Studio, Machine, Job
```

```python
studio = Studio(name="connor-pretrain", teamspace="medarc", org="medarc")
```

```python
jobs = {}

for jobid in range(2):
    job = Job.run(
        command=f"bash fmri-fm/experiments/pretrain_mix/launch_pretrain.sh {jobid}",
        name=f"pretrain_mix_{jobid}",
        machine=Machine.H100,
        studio=studio,
        interruptible=True,
    )
    print(job)
    jobs[jobid] = job
```
